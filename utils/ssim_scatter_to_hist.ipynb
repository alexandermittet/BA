```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for demonstration
data = np.random.normal(0.88, 0.04, 1000)

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, edgecolor='black', linewidth=0.5, density=True)

# Set the title and labels
plt.title('SSIM Scores for Validation Images', fontsize=14)
plt.xlabel('SSIM Score', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Set the x-axis and y-axis limits to match the scatterplot
plt.xlim(0.0, 1.0)


# Add grid lines
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for demonstration
data1 = np.random.normal(loc=0.9, scale=0.05, size=900)
data2 = np.random.normal(loc=0.65, scale=0.1, size=100)
data = np.concatenate((data1, data2))
data = data[(data >= 0) & (data <= 1)]

# Create a histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=50, edgecolor='black', linewidth=0.5, density=True)

# Set the title and labels
plt.title('SSIM Scores for Validation Images', fontsize=14)
plt.xlabel('SSIM Score', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)

# Set the x-axis and y-axis limits to match the scatterplot
plt.xlim(0.0, 1.0)
plt.ylim(0, 5.0)

# Add grid lines
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
