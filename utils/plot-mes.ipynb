```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os

# Read the DataFrame (assuming it's stored in a CSV file)


# Set the path to the directory containing the images
img_path = '/Users/alexandermittet/Library/CloudStorage/SeaDrive-almi(seafile.erda.dk) (14.03.2024 13.32)/My Libraries/BA_data/img'

df = pd.read_csv('/Users/alexandermittet/Documents/Seafile/ba-projekt/img_labels_ALL.csv')

# Filter the DataFrame to select one image for each score (0-4)
selected_images = []
for score in range(5):
    img_name = df[df['score'] == score]['img'].iloc[0]
    selected_images.append(img_name)

# Create a subplot for each image
fig, axs = plt.subplots(1, 5, figsize=(15, 3))

# Display the images in the subplots
for i, img_name in enumerate(selected_images):
    img_path_full = os.path.join(img_path, img_name)
    img = Image.open(img_path_full)
    axs[i].imshow(img)
    axs[i].set_title(f'Score: {i}')
    axs[i].axis('off')

# Adjust the spacing between subplots
plt.tight_layout()
plt.savefig('/Users/alexandermittet/Documents/Seafile/ba-projekt/plots/mes0to4.png')
# Display the plot
plt.show()
```
